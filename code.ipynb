```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
```

```python
train = pd.read_csv("Train.csv")
test = pd.read_csv("Test.csv")
test['Item_Outlet_Sales']=""
```

```python
train.head()
```

```python
train.dtypes
```

```python
categorical_variables = train.dtypes.loc[train.dtypes=='object'].index
print(categorical_variables)
```

```python
train[categorical_variables].apply(lambda x: len(x.unique()))
```

```python
continuous_variables = train.dtypes.loc[train.dtypes!='object'].index
print(continuous_variables)
```

```python
train.apply(lambda x: sum(x.isnull()))
```

```python
train["Outlet_Size"].fillna(train["Outlet_Size"].mode()[0],inplace = True)
test["Outlet_Size"].fillna(test["Outlet_Size"].mode()[0],inplace = True)
```

```python
train["Item_Weight"].fillna(train["Item_Weight"].mean(),inplace = True)
test["Item_Weight"].fillna(test["Item_Weight"].mean(),inplace = True)
```

```python
train.apply(lambda x: sum(x.isnull()))
```

```python
train.drop(["Outlet_Establishment_Year"],axis = 1)
test.drop(["Outlet_Establishment_Year"],axis = 1)
```

```python
train = pd.get_dummies(train,columns=[x for x in categorical_variables if x not in ['Item_Identifier','Outlet_Identifier']])
test = pd.get_dummies(test,columns=[x for x in categorical_variables if x not in ['Item_Identifier','Outlet_Identifier']])
train.head()
```

```python
dependent_variable = 'Item_Outlet_Sales'
independent_variables = [x for x in train.columns if x not in ['Item_Identifier','Outlet_Identifier',dependent_variable]]
print(independent_variables)
```

```python
scaler = StandardScaler()
#scaler.fit(train)
train[independent_variables] = scaler.fit_transform(train[independent_variables])
test[independent_variables] = scaler.fit_transform(test[independent_variables])
```

```python
model=RandomForestRegressor()
```

```python
model.fit(train[independent_variables],train[dependent_variable])
```

```python
test[dependent_variable]=model.predict(test[independent_variables])
test.head()
```

```python
print(model.score(train[independent_variables],train[dependent_variable]))
```

```python
submission = test[['Item_Identifier','Outlet_Identifier',dependent_variable]]
submission.head()
```

```python
submission.to_csv("SampleSubmission.csv")
```
